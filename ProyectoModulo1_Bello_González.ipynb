## 1. Entregable

### 1.1 Proyección de números ganadores para el sorteo nacional Melate.


### 1.2 Objetivos.

> #### 1.1 Objetivo general.
>  - El propósito de este trabajo es generar combinaciones de números ganadores del sorteo Melate

> #### 1.2 Objetivos específicos
>  - El propósito de este trabajo es analizar las combinaciones ganadoras del sorteo y apartr de ellas generar combinaciones con la más alta probabilidad de éxito para ganar el sorteo Melate



### 1.3 Definición del problema.
>Se desea saber los números mas probables para ganarse el sorteo 'Melate' para
obtener esta serie de números se simularán 10000 escenarios posibles.
El melate es un sorteo que se realizan todos los miércoles y domingos. Una
urna selecciona números al alzar los cuales pueden hacerte ganar.
La urna de Melate elegirá de manera aleatoria 7 esferas con los números
ganadores: los primeros 6 números seleccionados son llamados números naturales y el séptimo es el número adicional.

### 1.4 Nodos y variables que se decidieron simular y porqué
>Hemos identificado los siguientes escenarios a simular, ya que consideramos que
son sumamente importantes para llevar a cabo el proceso deseado:
  >  - Análisis de los números generados.
  >  - Generar números aletorios con la distribucion de los números ganadores.
  >  - Comprobar los números ganadores del proximo sorteo.
